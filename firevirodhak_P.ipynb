### Earth Engine Daily Ingestion & Per-Day Stack Export (Refactored)
This section ensures:
- Proper imports & EE init.
- Defines AOI (Pauri) and date range.
- Loads constant layers (DEM, Slope, Aspect, Hillshade, LULC) once.
- For each day builds a feature stack: Temperature (ERA5 daily mean), Wind (ERA5 10m u/v magnitude), NDVI (MODIS), Burn mask (MODIS MCD64A1 daily), plus constants.
- Stacks bands into one in-memory numpy array per day; optionally exports GeoTIFF (one multi-band file per date).
- Constants are duplicated per day intentionally.
Run this before temporal model training if raw GeoTIFF -> dynamic_array rebuild is needed.

```python
# Step 2: Clean feature stacks (NaN/inf handling, simple clipping)
import numpy as np
if 'feature_stacks' not in globals():
    print('feature_stacks missing; run Step 1')
else:
    fs = feature_stacks
    n_nan = np.isnan(fs).sum(); n_inf = np.isinf(fs).sum()
    if n_nan or n_inf:
        fs = np.nan_to_num(fs, nan=0.0, posinf=0.0, neginf=0.0)
    # Clip extreme values at 0.1 and 99.9 percentiles per band for stability
    T,F,H,W = fs.shape
    for b in range(F):
        band_vals = fs[:,b].ravel()
        lo, hi = np.percentile(band_vals, [0.1,99.9])
        fs[:,b] = np.clip(fs[:,b], lo, hi)
    feature_stacks = fs.astype('float32')
    print('Cleaned feature stacks; shape', feature_stacks.shape)
```

```python
import numpy as np, math, os, importlib
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from sklearn.metrics import (classification_report, confusion_matrix, roc_auc_score, average_precision_score, precision_recall_curve,
                             precision_recall_fscore_support)
from sklearn.isotonic import IsotonicRegression
from sklearn.model_selection import train_test_split

HISTORY_WINDOW_DAYS = 7
HORIZON_DAYS = 2
NEGATIVE_RATIO = 25
TARGET_RECALL = 0.60
RANDOM_STATE = 42
USE_LIGHTGBM = True

lgbm_available = importlib.util.find_spec('lightgbm') is not None if USE_LIGHTGBM else False
if lgbm_available:
    import lightgbm as lgb
else:
    from sklearn.ensemble import RandomForestClassifier
    print('LightGBM not available, falling back to RandomForest.')

if 'dynamic_array' not in globals() or 'constants_array' not in globals():
    raise RuntimeError('Run dataset prep first.')

T, Bdyn, H, W = dynamic_array.shape
assumed_band_order = ['LULC','DEM','Slope','Aspect','Hillshade','TempC','U10','V10','WindSpeed','NDVI','BurnDate']
const_bands = assumed_band_order[:5]
dyn_bands = assumed_band_order[5:]
idx_temp = dyn_bands.index('TempC'); idx_wind = dyn_bands.index('WindSpeed'); idx_ndvi = dyn_bands.index('NDVI'); idx_burn = dyn_bands.index('BurnDate')

try:
    doy_list = [datetime.strptime(d, '%Y-%m-%d').timetuple().tm_yday for d in dates]
except Exception:
    doy_list = list(range(1, T+1))
doy_arr = np.array(doy_list)

active_fire_cube = None
if 'ee' in globals():
    try:
        fire_ic = ee.ImageCollection('MODIS/061/MOD14A1').select('FireMask')
        af_arrays = []
        import rasterio
        with rasterio.open(stack_paths[0]) as src_demo:
            transform = src_demo.transform
        for dstr in dates:
            try:
                d = datetime.strptime(dstr, '%Y-%m-%d')
            except Exception:
                continue
            d_next = d + timedelta(days=1)
            subset = fire_ic.filterDate(dstr, d_next.strftime('%Y-%m-%d')).max().unmask(0)
            fire_bin = subset.gte(7).rename('AFire')
            fire_img = fire_bin.reproject(ee.Projection(CRS).atScale(SCALE)).clip(AOI)
            url = fire_img.getDownloadURL({'scale': SCALE, 'crs': CRS, 'region': AOI, 'format':'GEO_TIFF'})
            import requests, tempfile
            r = requests.get(url, stream=True, timeout=120)
            r.raise_for_status()
            with tempfile.NamedTemporaryFile(suffix='.tif', delete=False) as tmpf:
                for chunk in r.iter_content(8192):
                    if chunk: tmpf.write(chunk)
                tmp_path = tmpf.name
            with rasterio.open(tmp_path) as src:
                arr = src.read(1)
            os.remove(tmp_path)
            af_arrays.append(arr.astype(np.uint8))
        if af_arrays:
            active_fire_cube = np.stack(af_arrays, axis=0)
            if active_fire_cube.shape[0] != T:
                if active_fire_cube.shape[0] < T:
                    pad = np.zeros((T - active_fire_cube.shape[0], H, W), dtype=np.uint8)
                    active_fire_cube = np.concatenate([active_fire_cube, pad], axis=0)
            print('Active fire cube built with shape:', active_fire_cube.shape)
        else:
            print('No active fire detections retrieved (all zero).')
    except Exception as e:
        print('Active fire retrieval failed:', e)
        active_fire_cube = None
else:
    print('Earth Engine not available in this context, skipping active fire label enrichment.')

const_flat = constants_array.reshape(constants_array.shape[0], -1)
Npix = const_flat.shape[1]

temp_all = dynamic_array[:, idx_temp]
wind_all = dynamic_array[:, idx_wind]
ndvi_all = dynamic_array[:, idx_ndvi]
burn_all = dynamic_array[:, idx_burn]

HIST = HISTORY_WINDOW_DAYS
i_hist = np.arange(HIST, dtype=np.float32)
S_i = i_hist.sum(); S_i2 = (i_hist*i_hist).sum(); DEN = HIST*S_i2 - S_i**2 if (HIST*S_i2 - S_i**2)!=0 else 1.0

feat_blocks = []
label_blocks = []
day_blocks = []

for t in range(HIST, T):
    doy_t = doy_arr[t]
    win = slice(t-HIST, t)
    temp_hist = temp_all[win].reshape(HIST, -1)
    wind_hist = wind_all[win].reshape(HIST, -1)
    ndvi_hist = ndvi_all[win].reshape(HIST, -1)

    def stats(win_rs):
        last = win_rs[-1]; mean = win_rs.mean(0); std = win_rs.std(0)
        Sy = win_rs.sum(0); Siy = (win_rs * i_hist[:,None]).sum(0)
        slope = (HIST * Siy - S_i * Sy)/DEN
        return last, mean, std, slope

    t_last, t_mean, t_std, t_slope = stats(temp_hist)
    w_last, w_mean, w_std, w_slope = stats(wind_hist)
    n_last, n_mean, n_std, n_slope = stats(ndvi_hist)

    t_anom = t_last - t_mean
    n_anom = n_last - n_mean
    dryness = (1 - (n_last + 1)/2).clip(0,1)
    hot_days = (temp_hist > 30.0).sum(0).astype(np.float32)
    windy_days = (wind_hist > 5.0).sum(0).astype(np.float32)

    burn_doy = burn_all[t].reshape(-1)
    mask_not_burned = (burn_doy == 0) | (burn_doy >= doy_t)

    burn_future = (burn_doy > doy_t) & (burn_doy <= doy_t + HORIZON_DAYS)

    if active_fire_cube is not None and active_fire_cube.shape[0] == T:
        af_future = np.zeros_like(burn_future)
        for dt in range(1, HORIZON_DAYS+1):
            if t+dt < T:
                af_future |= (active_fire_cube[t+dt] > 0).reshape(-1)
    else:
        af_future = np.zeros_like(burn_future)

    refined_future = ( (burn_future | af_future) & mask_not_burned )

    keep_idx = np.where(mask_not_burned)[0]
    if keep_idx.size == 0:
        continue

    days_since = np.where((burn_doy > 0) & (burn_doy < doy_t), doy_t - burn_doy, 0).astype(np.float32)
    sin_doy = math.sin(2*math.pi*doy_t/365.0); cos_doy = math.cos(2*math.pi*doy_t/365.0)
    sin_feat = np.full(keep_idx.size, sin_doy, dtype=np.float32)
    cos_feat = np.full(keep_idx.size, cos_doy, dtype=np.float32)

    stacked = np.vstack([
        const_flat[:, keep_idx],
        t_last[keep_idx], t_mean[keep_idx], t_std[keep_idx], t_slope[keep_idx], t_anom[keep_idx],
        w_last[keep_idx], w_mean[keep_idx], w_std[keep_idx], w_slope[keep_idx],
        n_last[keep_idx], n_mean[keep_idx], n_std[keep_idx], n_slope[keep_idx], n_anom[keep_idx], dryness[keep_idx],
        hot_days[keep_idx], windy_days[keep_idx], days_since[keep_idx],
        sin_feat, cos_feat
    ])

    feat_blocks.append(stacked.T)
    label_blocks.append(refined_future[keep_idx].astype(np.uint8))
    day_blocks.append(np.full(keep_idx.size, t, dtype=np.int32))

if not feat_blocks:
    raise RuntimeError('No samples generated with refined labels.')

X_all = np.concatenate(feat_blocks, axis=0)
y_all = np.concatenate(label_blocks, axis=0)
day_idx_all = np.concatenate(day_blocks, axis=0)
print(f'Refined samples: {X_all.shape} Positives: {int(y_all.sum())} Pos%={100*y_all.mean():.4f}')

if y_all.sum() == 0:
    print('Still zero positives after label enrichment. Increase HORIZON_DAYS or verify active fire retrieval.')

POS = np.where(y_all==1)[0]
NEG = np.where(y_all==0)[0]
max_neg = min(len(NEG), NEGATIVE_RATIO * len(POS)) if len(POS)>0 else 0
if max_neg and max_neg < len(NEG):
    sel_neg = np.random.choice(NEG, max_neg, replace=False)
    sel_idx = np.concatenate([POS, sel_neg])
    np.random.shuffle(sel_idx)
    X_bal = X_all[sel_idx]
    y_bal = y_all[sel_idx]
    day_bal = day_idx_all[sel_idx]
    print(f'Downsampled negatives {len(NEG)} -> {max_neg}; New shape {X_bal.shape}')
else:
    X_bal, y_bal, day_bal = X_all, y_all, day_idx_all

u_pos_days, pos_counts = np.unique(day_bal[y_bal==1], return_counts=True) if y_bal.sum()>0 else ([],[])
if len(u_pos_days) > 1:
    cumulative = 0; desired_test_pos = max(1, int(0.2 * y_bal.sum())); split_candidate = None
    for d, c in sorted(zip(u_pos_days, pos_counts)):
        cumulative += c
        if y_bal.sum() - cumulative >= desired_test_pos:
            split_candidate = d + 1
        else:
            break
    if split_candidate is None:
        split_candidate = u_pos_days[-2] + 1 if len(u_pos_days)>1 else u_pos_days[0]
    train_mask = day_bal < split_candidate; test_mask = day_bal >= split_candidate
else:
    train_mask = np.zeros(len(y_bal), dtype=bool)
    train_mask[:int(0.7*len(y_bal))] = True
    np.random.shuffle(train_mask)
    test_mask = ~train_mask

X_train, y_train = X_bal[train_mask], y_bal[train_mask]
X_test, y_test = X_bal[test_mask], y_bal[test_mask]
print('Train/Test shapes:', X_train.shape, X_test.shape, '| Pos train:', y_train.sum(), 'Pos test:', y_test.sum())

if lgbm_available:
    lgb_train = lgb.Dataset(X_train, label=y_train)
    params = {
        'objective':'binary',
        'learning_rate':0.05,
        'num_leaves':32,
        'feature_fraction':0.8,
        'bagging_fraction':0.8,
        'bagging_freq':1,
        'min_data_in_leaf':20,
        'metric':['auc'],
        'verbose':-1,
        'seed':RANDOM_STATE,
        'scale_pos_weight': (len(y_train)-y_train.sum())/max(1,y_train.sum()) if y_train.sum()>0 else 1.0
    }
    clf_ref = lgb.train(params, lgb_train, num_boost_round=400)
    y_prob = clf_ref.predict(X_test)
else:
    clf_ref = RandomForestClassifier(n_estimators=600, max_depth=24, min_samples_leaf=2, n_jobs=-1, random_state=RANDOM_STATE, class_weight='balanced_subsample')
    clf_ref.fit(X_train, y_train)
    y_prob = clf_ref.predict_proba(X_test)[:,1]

if y_train.sum() >= 50:
    X_tr, X_cal, y_tr, y_cal = train_test_split(X_train, y_train, test_size=0.2, stratify=y_train, random_state=RANDOM_STATE)
    if lgbm_available:
        clf_ref = lgb.train(params, lgb.Dataset(X_tr, label=y_tr), num_boost_round=400)
        cal_prob = clf_ref.predict(X_cal)
    else:
        clf_ref.fit(X_tr, y_tr)
        cal_prob = clf_ref.predict_proba(X_cal)[:,1]
    iso = IsotonicRegression(out_of_bounds='clip')
    iso.fit(cal_prob, y_cal)
    if lgbm_available:
        clf_ref = lgb.train(params, lgb.Dataset(X_train, label=y_train), num_boost_round=400)
        y_prob_raw = clf_ref.predict(X_test)
    else:
        clf_ref.fit(X_train, y_train)
        y_prob_raw = clf_ref.predict_proba(X_test)[:,1]
    y_prob = iso.transform(y_prob_raw)
    print('Applied isotonic calibration.')

y_pred_default = (y_prob >= 0.5).astype(int)
print('\nBaseline threshold (0.5) classification report:')
print(classification_report(y_test, y_pred_default, zero_division=0))
print('Confusion matrix:\n', confusion_matrix(y_test, y_pred_default))
try:
    roc = roc_auc_score(y_test, y_prob); pr_auc = average_precision_score(y_test, y_prob)
    print(f'ROC AUC: {roc:.4f} | PR AUC: {pr_auc:.4f}')
except Exception:
    pass

precisions, recalls, thr = precision_recall_curve(y_test, y_prob)
f1_scores = 2*precisions[:-1]*recalls[:-1]/(precisions[:-1]+recalls[:-1]+1e-9)
best_idx = np.argmax(f1_scores); best_thr = thr[best_idx]
print(f'Best F1 threshold: {best_thr:.4f} (Prec={precisions[best_idx]:.4f} Rec={recalls[best_idx]:.4f} F1={f1_scores[best_idx]:.4f})')
rec_target_indices = np.where(recalls[:-1] >= TARGET_RECALL)[0]
if rec_target_indices.size:
    cand_idx = rec_target_indices[np.argmax(precisions[rec_target_indices])]
    rec_thr = thr[cand_idx]
    print(f'Target recall {TARGET_RECALL:.2f} threshold: {rec_thr:.4f} (Prec={precisions[cand_idx]:.4f} Rec={recalls[cand_idx]:.4f})')
else:
    rec_thr = None
    print(f'No threshold achieves recall >= {TARGET_RECALL:.2f}.')

y_pred_best = (y_prob >= best_thr).astype(int)
prec_b, rec_b, f1_b, _ = precision_recall_fscore_support(y_test, y_pred_best, average='binary', zero_division=0)
print(f'Performance @ best F1 thr: Prec={prec_b:.4f} Rec={rec_b:.4f} F1={f1_b:.4f}')
if rec_thr is not None:
    y_pred_rec = (y_prob >= rec_thr).astype(int)
    prec_r, rec_r, f1_r, _ = precision_recall_fscore_support(y_test, y_pred_rec, average='binary', zero_division=0)
    print(f'Performance @ target recall thr: Prec={prec_r:.4f} Rec={rec_r:.4f} F1={f1_r:.4f}')

import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.plot(recalls, precisions); plt.xlabel('Recall'); plt.ylabel('Precision'); plt.title('PR Curve')
plt.scatter(recalls[best_idx], precisions[best_idx], c='red', s=30, label='Best F1')
plt.legend(); plt.grid(alpha=0.3)
plt.subplot(1,3,2)
plt.plot(thr, f1_scores); plt.axvline(best_thr, color='red', linestyle='--'); plt.xlabel('Threshold'); plt.ylabel('F1'); plt.title('F1 vs Threshold'); plt.grid(alpha=0.3)
plt.subplot(1,3,3)
plt.hist(y_prob[y_test==0], bins=40, alpha=0.6, label='Neg'); plt.hist(y_prob[y_test==1], bins=40, alpha=0.6, label='Pos')
plt.axvline(best_thr, color='red', linestyle='--');
if rec_thr is not None: plt.axvline(rec_thr, color='orange', linestyle=':')
plt.title('Probability Distribution'); plt.legend(); plt.tight_layout(); plt.show()

print('\nNEXT STEPS: If metrics still low, expand feature set (precip, humidity, multi-year NDVI anomalies) and re-run.')
```

```python
import numpy as np, math, os, importlib
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from sklearn.metrics import (classification_report, confusion_matrix, roc_auc_score, average_precision_score, precision_recall_curve, precision_recall_fscore_support)
from sklearn.model_selection import train_test_split

if 'MOISTURE_RUN_DONE' in globals() and not globals().get('FORCE_RERUN_MOISTURE', False):
    print('Moisture feature pipeline already executed. Set FORCE_RERUN_MOISTURE=True to recompute.')
else:
    HISTORY_WINDOW_DAYS = 7
    HORIZON_DAYS = 2
    NEGATIVE_RATIO = 25
    RANDOM_STATE = 42
    TARGET_RECALL = 0.60
    USE_LIGHTGBM = True

    lgbm_available = importlib.util.find_spec('lightgbm') is not None if USE_LIGHTGBM else False
    if lgbm_available:
        import lightgbm as lgb
    else:
        from sklearn.ensemble import RandomForestClassifier
        print('LightGBM not available; fallback to RandomForest.')

    if 'dynamic_array' not in globals():
        raise RuntimeError('Run earlier cells to build dynamic_array.')

    precip_cube = None
    rh_cube = None
    vpd_cube = None

    if 'ee' in globals():
        try:
            era5 = ee.ImageCollection('ECMWF/ERA5/DAILY')
            precip_arrays = []
            rh_arrays = []
            vpd_arrays = []
            for dstr in dates:
                d = datetime.strptime(dstr, '%Y-%m-%d')
                d_next = d + timedelta(days=1)
                date_filter = ee.Filter.date(dstr, d_next.strftime('%Y-%m-%d'))
                img = era5.filter(date_filter).first()
                if img is None:
                    precip_arrays.append(np.zeros((dynamic_array.shape[2], dynamic_array.shape[3]), dtype=np.float32))
                    rh_arrays.append(np.zeros_like(precip_arrays[-1]))
                    vpd_arrays.append(np.zeros_like(precip_arrays[-1]))
                    continue
                precip = ee.Image(img).select('total_precipitation').multiply(1000).rename('Precip_mm')
                t2m = ee.Image(img).select('mean_2m_air_temperature')
                dew = ee.Image(img).select('dewpoint_2m_temperature')
                t2c = t2m.add(-273.15)
                dew_c = dew.add(-273.15)
                es = t2c.multiply(17.27).divide(t2c.add(237.3)).exp().multiply(0.6108)
                ea = dew_c.multiply(17.27).divide(dew_c.add(237.3)).exp().multiply(0.6108)
                rh = ea.divide(es).multiply(100).rename('RH')
                vpd = es.subtract(ea).rename('VPD')
                combo = precip.addBands(rh).addBands(vpd).reproject(ee.Projection(CRS).atScale(SCALE)).clip(AOI)
                import requests, tempfile, rasterio
                url = combo.getDownloadURL({'scale': SCALE, 'crs': CRS, 'region': AOI, 'format':'GEO_TIFF'})
                r = requests.get(url, stream=True, timeout=120); r.raise_for_status()
                with tempfile.NamedTemporaryFile(suffix='.tif', delete=False) as tmpf:
                    for chunk in r.iter_content(8192):
                        if chunk: tmpf.write(chunk)
                    tmp_path = tmpf.name
                with rasterio.open(tmp_path) as src:
                    arrs = src.read()
                os.remove(tmp_path)
                precip_arrays.append(arrs[0].astype(np.float32))
                rh_arrays.append(arrs[1].astype(np.float32))
                vpd_arrays.append(arrs[2].astype(np.float32))
            precip_cube = np.stack(precip_arrays, axis=0)
            rh_cube = np.stack(rh_arrays, axis=0)
            vpd_cube = np.stack(vpd_arrays, axis=0)
            print('Retrieved extra ERA5 cubes shapes:', precip_cube.shape, rh_cube.shape, vpd_cube.shape)
        except Exception as e:
            print('Moisture feature retrieval failed:', e)
    else:
        print('Earth Engine not available; skipping moisture features.')

    if precip_cube is None:
        precip_cube = np.zeros((dynamic_array.shape[0], dynamic_array.shape[2], dynamic_array.shape[3]), dtype=np.float32)
    if rh_cube is None:
        rh_cube = np.zeros_like(precip_cube)
    if vpd_cube is None:
        vpd_cube = np.zeros_like(precip_cube)

    assumed_band_order = ['LULC','DEM','Slope','Aspect','Hillshade','TempC','U10','V10','WindSpeed','NDVI','BurnDate']
    const_bands = assumed_band_order[:5]
    dyn_bands = assumed_band_order[5:]
    idx_temp = dyn_bands.index('TempC'); idx_wind = dyn_bands.index('WindSpeed'); idx_ndvi = dyn_bands.index('NDVI'); idx_burn = dyn_bands.index('BurnDate')

    try:
        doy_list = [datetime.strptime(d, '%Y-%m-%d').timetuple().tm_yday for d in dates]
    except Exception:
        doy_list = list(range(1, dynamic_array.shape[0]+1))
    doy_arr = np.array(doy_list)

    const_flat = constants_array.reshape(constants_array.shape[0], -1)
    Npix = const_flat.shape[1]

    temp_all = dynamic_array[:, idx_temp]
    wind_all = dynamic_array[:, idx_wind]
    ndvi_all = dynamic_array[:, idx_ndvi]
    burn_all = dynamic_array[:, idx_burn]

    HIST = HISTORY_WINDOW_DAYS
    i_hist = np.arange(HIST, dtype=np.float32)
    S_i = i_hist.sum(); S_i2 = (i_hist*i_hist).sum(); DEN = HIST*S_i2 - S_i**2 if (HIST*S_i2 - S_i**2)!=0 else 1.0

    feature_blocks = []
    label_blocks = []

    for t in range(HIST, dynamic_array.shape[0]):
        doy_t = doy_arr[t]
        win = slice(t-HIST, t)
        temp_hist = temp_all[win].reshape(HIST, -1)
        wind_hist = wind_all[win].reshape(HIST, -1)
        ndvi_hist = ndvi_all[win].reshape(HIST, -1)
        precip_hist = precip_cube[win].reshape(HIST, -1)
        rh_hist = rh_cube[win].reshape(HIST, -1)
        vpd_hist = vpd_cube[win].reshape(HIST, -1)

        def stats(win_rs):
            last = win_rs[-1]; mean = win_rs.mean(0); std = win_rs.std(0)
            Sy = win_rs.sum(0); Siy = (win_rs * i_hist[:,None]).sum(0)
            slope = (HIST * Siy - S_i * Sy)/DEN
            return last, mean, std, slope

        t_last, t_mean, t_std, t_slope = stats(temp_hist)
        w_last, w_mean, w_std, w_slope = stats(wind_hist)
        n_last, n_mean, n_std, n_slope = stats(ndvi_hist)
        p_last, p_mean, p_std, p_slope = stats(precip_hist)
        rh_last, rh_mean, _, _ = stats(rh_hist)
        vpd_last, vpd_mean, _, _ = stats(vpd_hist)

        precip_sum = precip_hist.sum(0)
        dry_day_count = (precip_hist < 0.5).sum(0).astype(np.float32)

        t_anom = t_last - t_mean
        n_anom = n_last - n_mean
        dryness = (1 - (n_last + 1)/2).clip(0,1)
        hot_days = (temp_hist > 30.0).sum(0).astype(np.float32)
        windy_days = (wind_hist > 5.0).sum(0).astype(np.float32)

        burn_doy = burn_all[t].reshape(-1)
        mask_not_burned = (burn_doy == 0) | (burn_doy >= doy_t)
        burn_future = (burn_doy > doy_t) & (burn_doy <= doy_t + HORIZON_DAYS)

        keep_idx = np.where(mask_not_burned)[0]
        if keep_idx.size == 0:
            continue

        days_since = np.where((burn_doy > 0) & (burn_doy < doy_t), doy_t - burn_doy, 0).astype(np.float32)
        sin_doy = math.sin(2*math.pi*doy_t/365.0); cos_doy = math.cos(2*math.pi*doy_t/365.0)
        sin_feat = np.full(keep_idx.size, sin_doy, dtype=np.float32)
        cos_feat = np.full(keep_idx.size, cos_doy, dtype=np.float32)

        if 'active_fire_cube' in globals() and active_fire_cube is not None and active_fire_cube.shape[0] == dynamic_array.shape[0]:
            af_future = np.zeros_like(burn_future)
            for dt in range(1, HORIZON_DAYS+1):
                if t+dt < active_fire_cube.shape[0]:
                    af_future |= active_fire_cube[t+dt].reshape(-1) > 0
            label_future = ( (burn_future | af_future) & mask_not_burned )
        else:
            label_future = (burn_future & mask_not_burned)

        stacked = np.vstack([
            const_flat[:, keep_idx],
            t_last[keep_idx], t_mean[keep_idx], t_std[keep_idx], t_slope[keep_idx], t_anom[keep_idx],
            w_last[keep_idx], w_mean[keep_idx], w_std[keep_idx], w_slope[keep_idx],
            n_last[keep_idx], n_mean[keep_idx], n_std[keep_idx], n_slope[keep_idx], n_anom[keep_idx], dryness[keep_idx],
            hot_days[keep_idx], windy_days[keep_idx], days_since[keep_idx],
            p_last[keep_idx], p_mean[keep_idx], p_std[keep_idx], p_slope[keep_idx], precip_sum[keep_idx], dry_day_count[keep_idx],
            rh_last[keep_idx], rh_mean[keep_idx], vpd_last[keep_idx], vpd_mean[keep_idx],
            sin_feat, cos_feat,
            dry_day_count[keep_idx]
        ])

        feature_blocks.append(stacked.T)
        label_blocks.append(label_future[keep_idx].astype(np.uint8))

    if not feature_blocks:
        raise RuntimeError('No samples generated with moisture features.')

    X_all = np.concatenate(feature_blocks, axis=0)
    y_all = np.concatenate(label_blocks, axis=0)
    print('Moisture feature sample matrix:', X_all.shape, 'Positives:', int(y_all.sum()), f'Pos%={100*y_all.mean():.4f}')
    if y_all.sum()==0:
        print('No positives present; cannot train.')
    POS = np.where(y_all==1)[0]; NEG = np.where(y_all==0)[0]
    max_neg = min(len(NEG), NEGATIVE_RATIO * len(POS)) if len(POS)>0 else 0
    if max_neg and max_neg < len(NEG):
        sel_neg = np.random.choice(NEG, max_neg, replace=False)
        sel_idx = np.concatenate([POS, sel_neg]); np.random.shuffle(sel_idx)
        X_bal = X_all[sel_idx]; y_bal = y_all[sel_idx]
        print(f'Negatives downsampled {len(NEG)} -> {max_neg}; New shape {X_bal.shape}')
    else:
        X_bal, y_bal = X_all, y_all

    if y_bal.sum() > 0 and y_bal.sum() < len(y_bal):
        X_train, X_test, y_train, y_test = train_test_split(X_bal, y_bal, test_size=0.25, stratify=y_bal, random_state=RANDOM_STATE)
    else:
        X_train, X_test, y_train, y_test = X_bal, X_bal, y_bal, y_bal
    print('Train/Test:', X_train.shape, X_test.shape, '| Pos train:', y_train.sum(), 'Pos test:', y_test.sum())

    if lgbm_available:
        lgb_train = lgb.Dataset(X_train, label=y_train)
        params = {
            'objective':'binary', 'learning_rate':0.05, 'num_leaves':48, 'feature_fraction':0.85,
            'bagging_fraction':0.8, 'bagging_freq':1, 'min_data_in_leaf':25, 'metric':['auc'], 'verbose':-1,
            'seed':RANDOM_STATE,
            'scale_pos_weight': (len(y_train)-y_train.sum())/max(1,y_train.sum()) if y_train.sum()>0 else 1.0
        }
        clf_m = lgb.train(params, lgb_train, num_boost_round=500)
        y_prob = clf_m.predict(X_test)
    else:
        clf_m = RandomForestClassifier(n_estimators=700, max_depth=28, min_samples_leaf=2, n_jobs=-1, random_state=RANDOM_STATE, class_weight='balanced_subsample')
        clf_m.fit(X_train, y_train)
        y_prob = clf_m.predict_proba(X_test)[:,1]

    if y_test.sum()>0:
        print('\nBaseline 0.5 threshold:')
        y_pred = (y_prob >= 0.5).astype(int)
        print(classification_report(y_test, y_pred, zero_division=0))
        print('Confusion matrix:\n', confusion_matrix(y_test, y_pred))
        try:
            roc = roc_auc_score(y_test, y_prob); pr_auc = average_precision_score(y_test, y_prob)
            print(f'ROC AUC: {roc:.4f} | PR AUC: {pr_auc:.4f}')
        except Exception:
            pass
        precisions, recalls, thrs = precision_recall_curve(y_test, y_prob)
        f1s = 2*precisions[:-1]*recalls[:-1]/(precisions[:-1]+recalls[:-1]+1e-9)
        best_idx = np.argmax(f1s); best_thr = thrs[best_idx]
        print(f'Best F1 threshold: {best_thr:.4f} (Prec={precisions[best_idx]:.4f} Rec={recalls[best_idx]:.4f} F1={f1s[best_idx]:.4f})')
        rec_target_idxs = np.where(recalls[:-1] >= TARGET_RECALL)[0]
        if rec_target_idxs.size:
            ridx = rec_target_idxs[np.argmax(precisions[rec_target_idxs])]
            rec_thr = thrs[ridx]
            print(f'Target recall {TARGET_RECALL:.2f} threshold: {rec_thr:.4f} (Prec={precisions[ridx]:.4f} Rec={recalls[ridx]:.4f})')
        else:
            rec_thr = None
        y_pred_best = (y_prob >= best_thr).astype(int)
        pb, rb, f1b, _ = precision_recall_fscore_support(y_test, y_pred_best, average='binary', zero_division=0)
        print(f'Performance @ best F1 thr: Prec={pb:.4f} Rec={rb:.4f} F1={f1b:.4f}')
        if rec_thr is not None:
            y_pred_rec = (y_prob >= rec_thr).astype(int)
            prc, rrc, f1c, _ = precision_recall_fscore_support(y_test, y_pred_rec, average='binary', zero_division=0)
            print(f'Performance @ target recall thr: Prec={prc:.4f} Rec={rrc:.4f} F1={f1c:.4f}')
        plt.figure(figsize=(12,4))
        plt.subplot(1,3,1); plt.plot(recalls, precisions); plt.scatter(recalls[best_idx], precisions[best_idx], c='red'); plt.xlabel('Recall'); plt.ylabel('Precision'); plt.title('PR Curve'); plt.grid(alpha=0.3)
        plt.subplot(1,3,2); plt.plot(thrs, f1s); plt.axvline(best_thr, color='red', ls='--'); plt.xlabel('Threshold'); plt.ylabel('F1'); plt.title('F1 vs Threshold'); plt.grid(alpha=0.3)
        plt.subplot(1,3,3); plt.hist(y_prob[y_test==0], bins=40, alpha=0.6, label='Neg'); plt.hist(y_prob[y_test==1], bins=40, alpha=0.6, label='Pos'); plt.axvline(best_thr, color='red', ls='--');
        if rec_thr is not None: plt.axvline(rec_thr, color='orange', ls=':'); plt.legend(); plt.title('Probability Distribution'); plt.tight_layout(); plt.show()
    else:
        print('No positives in test set; cannot evaluate metrics.')

    MOISTURE_RUN_DONE = True
    print('Done moisture feature retraining (updated dewpoint band).')
```

```python
import numpy as np, os, io, math
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from matplotlib.colors import ListedColormap
from matplotlib import patheffects

try:
    from PIL import Image
    pil_ok = True
except ImportError:
    pil_ok = False

try:
    import imageio
    have_imageio = True
except ImportError:
    have_imageio = False

FORECAST_DAYS_TO_SHOW = 5
USE_THRESHOLD = None
SMOOTH_SIGMA = 0.6
FEEDBACK_BURN = True
BURN_FEEDBACK_MIN_PROB = 0.65
PROB_POWER = 1.2
SHOW_UNCERTAINTY = True
SAVE_DAILY_GEOTIFFS = False
WRITE_MP4 = False
USE_ADAPTIVE_THRESHOLD = True
ADAPTIVE_TOP_FRACTION = 0.005
ADAPTIVE_MIN_PIXELS = 50
OUT_GIF = os.path.join(OUTPUT_DIR, 'enhanced_future_fire_spread.gif')
OUT_MP4 = os.path.join(OUTPUT_DIR, 'enhanced_future_fire_spread.mp4')
GEOTIFF_DIR = os.path.join(OUTPUT_DIR, 'forecast_probs')
DEBUG = True
FRAME_DURATION_SEC = globals().get('FRAME_DURATION_SEC', 0.8)

if 'dynamic_array' not in globals() or 'constants_array' not in globals():
    print('Required arrays missing; run earlier cells.')
else:
    model = None; lightgbm_mode=False; rf_mode=False
    if 'clf_m' in globals() and clf_m is not None:
        model = clf_m; lightgbm_mode=True
    elif 'clf' in globals() and clf is not None:
        model = clf
        try:
            from sklearn.ensemble import RandomForestClassifier
            rf_mode = isinstance(model, RandomForestClassifier)
        except Exception:
            rf_mode = False
    else:
        print('No trained model found.');

    def pad_align_matrix(Xm):
        """Ensure feature matrix matches LightGBM expected feature count (pad or trim)."""
        if not lightgbm_mode or model is None: return Xm
        try:
            m_feats = list(model.feature_name())
            expected = len(m_feats)
        except Exception:
            return Xm
        cur = Xm.shape[1]
        if cur == expected:
            return Xm
        if cur < expected:
            if DEBUG:
                print(f'Padding feature matrix from {cur} -> {expected}')
            pad = np.zeros((Xm.shape[0], expected-cur), dtype=Xm.dtype)
            return np.hstack([Xm, pad])
        if cur > expected:
            if DEBUG:
                print(f'Trimming feature matrix from {cur} -> {expected}')
            return Xm[:, :expected]
        return Xm

    if model is not None:
        thr = float(BEST_F1_THRESHOLD) if USE_THRESHOLD is None and 'BEST_F1_THRESHOLD' in globals() else (USE_THRESHOLD if USE_THRESHOLD is not None else 0.5)
        print(f'Using base threshold: {thr:.4f}')
        if 'feat_names' not in globals() or len(feat_names)!=35:
            if 'FULL_FEATURE_ORDER' in globals():
                feat_names = FULL_FEATURE_ORDER.copy()
            else:
                print('Warning: feat_names length mismatch and fallback order missing.')
        canonical_order = feat_names
        if lightgbm_mode and hasattr(model,'feature_name'):
            model_order = list(model.feature_name())
        else:
            model_order = canonical_order
        positional = all(n.startswith('Column_') for n in model_order)

        # If model expects more features than canonical, extend canonical placeholders
        if lightgbm_mode and len(model_order) != len(canonical_order):
            missing_model_feats = [f for f in model_order if f not in canonical_order and not f.startswith('Column_')]
            if missing_model_feats:
                print('Model has extra features not in canonical list:', missing_model_feats)
            # No change to feature construction logic; padding handled later.

        Tcur,Ddyn,Hh,Ww = dynamic_array.shape
        hist = HISTORY_WINDOW_DAYS if 'HISTORY_WINDOW_DAYS' in globals() else 7
        base_date = datetime.strptime(dates[-1], '%Y-%m-%d')

        have_precip = 'precip_cube' in globals() and isinstance(precip_cube, np.ndarray) and precip_cube.shape[0]==Tcur
        have_rh = 'rh_cube' in globals() and isinstance(rh_cube, np.ndarray) and rh_cube.shape[0]==Tcur
        have_vpd = 'vpd_cube' in globals() and isinstance(vpd_cube, np.ndarray) and vpd_cube.shape[0]==Tcur
        have_dryness = 'dryness' in globals() and isinstance(dryness, np.ndarray) and dryness.shape[0]==Tcur
        have_dry_day_count = 'dry_day_count' in globals() and dry_day_count.shape[0]==Tcur
        have_hot_days = 'hot_days' in globals() and hot_days.shape[0]==Tcur
        have_windy_days = 'windy_days' in globals() and windy_days.shape[0]==Tcur

        precip_future = precip_cube.copy() if have_precip else None
        rh_future = rh_cube.copy() if have_rh else None
        vpd_future = vpd_cube.copy() if have_vpd else None
        dryness_future = dryness.copy() if have_dryness else None
        dry_day_count_future = dry_day_count.copy() if have_dry_day_count else None
        hot_days_future = hot_days.copy() if have_hot_days else None
        windy_days_future = windy_days.copy() if have_windy_days else None

        last_dyn = dynamic_array[-1].copy()
        past_dyn = dynamic_array.copy()

        if SAVE_DAILY_GEOTIFFS:
            os.makedirs(GEOTIFF_DIR, exist_ok=True)
            try:
                import rasterio
                have_rasterio = True
            except ImportError:
                have_rasterio = False
                print('GeoTIFF save requested but rasterio not installed.')
        else:
            have_rasterio = False

        def smooth(a,s):
            if not s or s<=0: return a
            try:
                from scipy.ndimage import gaussian_filter
                return gaussian_filter(a, sigma=float(s))
            except Exception:
                return a

        def stats(series):
            last=series[-1]; mean=series.mean(0); std=series.std(0)
            idx=np.arange(series.shape[0]); t_mean=idx.mean(); x_mean=series.mean(0)
            num=np.sum((idx[:,None,None]-t_mean)*(series-x_mean), axis=0); den=np.sum((idx-t_mean)**2)
            slope = num/den if den>0 else np.zeros_like(mean)
            return last,mean,std,slope

        def build_canonical_feature_stack(past_dyn, step_offset):
            window = past_dyn[-hist:]
            temp=window[:,0]; wind=window[:,3]; ndvi=window[:,4]; burn=window[:,5]
            t_last,t_mean,t_std,t_slope = stats(temp)
            w_last,w_mean,w_std,w_slope = stats(wind)
            n_last,n_mean,n_std,n_slope = stats(ndvi)
            temp_anom = t_last - t_mean
            ndvi_anom = n_last - n_mean
            burned_prior = (burn[:-1]>0).any(axis=0)
            days_since_burn = np.where(burned_prior, hist, 0).astype(np.float32)
            if precip_future is not None:
                p_window = precip_future[-hist:]
                p_last,p_mean,p_std,p_slope = stats(p_window)
                p_sum = p_window.sum(0)
            else:
                shape=t_last.shape; p_last=p_mean=p_std=p_slope=p_sum=np.zeros(shape,np.float32)
            if rh_future is not None:
                rh_window=rh_future[-hist:]; rh_last,rh_mean,_,_ = stats(rh_window)
            else:
                rh_last=rh_mean=np.zeros_like(t_last)
            if vpd_future is not None:
                vpd_window=vpd_future[-hist:]; vpd_last,vpd_mean,_,_ = stats(vpd_window)
            else:
                vpd_last=vpd_mean=np.zeros_like(t_last)
            dryness_val = dryness_future[-1] if dryness_future is not None else np.zeros_like(t_last)
            hot_ct = hot_days_future[-1] if hot_days_future is not None else np.zeros_like(t_last)
            windy_ct = windy_days_future[-1] if windy_days_future is not None else np.zeros_like(t_last)
            dry_day_ct = dry_day_count_future[-1] if dry_day_count_future is not None else np.zeros_like(t_last)
            fut_date = base_date + timedelta(days=step_offset)
            doy = fut_date.timetuple().tm_yday
            sin_doy = math.sin(2*math.pi*doy/365.0); cos_doy = math.cos(2*math.pi*doy/365.0)
            sin_doy_arr = np.full_like(t_last, sin_doy, dtype=np.float32)
            cos_doy_arr = np.full_like(t_last, cos_doy, dtype=np.float32)
            consts = list(constants_array)
            const_map = {
                'CONST_LULC': consts[0] if len(consts)>0 else np.zeros_like(t_last),
                'CONST_DEM': consts[1] if len(consts)>1 else np.zeros_like(t_last),
                'CONST_Slope': consts[2] if len(consts)>2 else np.zeros_like(t_last),
                'CONST_Aspect': consts[3] if len(consts)>3 else np.zeros_like(t_last),
                'CONST_Hillshade': consts[4] if len(consts)>4 else np.zeros_like(t_last),
            }
            feat_dict = {
                'CONST_LULC': const_map['CONST_LULC'], 'CONST_DEM': const_map['CONST_DEM'], 'CONST_Slope': const_map['CONST_Slope'], 'CONST_Aspect': const_map['CONST_Aspect'], 'CONST_Hillshade': const_map['CONST_Hillshade'],
                'Temp_last': t_last,'Temp_mean': t_mean,'Temp_std': t_std,'Temp_slope': t_slope,'Temp_anom': temp_anom,
                'Wind_last': w_last,'Wind_mean': w_mean,'Wind_std': w_std,'Wind_slope': w_slope,
                'NDVI_last': n_last,'NDVI_mean': n_mean,'NDVI_std': n_std,'NDVI_slope': n_slope,'NDVI_anom': ndvi_anom,
                'Dryness': dryness_val,'HotDayCount': hot_ct,'WindyDayCount': windy_ct,'DaysSinceBurn': days_since_burn,'DOY_sin': sin_doy_arr,'DOY_cos': cos_doy_arr,
                'Precip_last': p_last,'Precip_mean': p_mean,'Precip_std': p_std,'Precip_slope': p_slope,'Precip_sum': p_sum,
                'RH_last': rh_last,'RH_mean': rh_mean,'VPD_last': vpd_last,'VPD_mean': vpd_mean,'DryDayCount': dry_day_ct
            }
            stack = [feat_dict.get(n, np.zeros_like(t_last)) for n in canonical_order]
            return np.stack(stack, axis=0)

        def to_model_matrix(feat_stack):
            if positional and len(model_order)==feat_stack.shape[0]:
                ordered = feat_stack
            else:
                name_arr = {canonical_order[i]: feat_stack[i] for i in range(len(canonical_order))}
                ordered_list=[]
                for mo in model_order:
                    if mo in name_arr:
                        ordered_list.append(name_arr[mo])
                    elif mo.startswith('Column_'):
                        idx=int(mo.split('_')[1])
                        ordered_list.append(feat_stack[idx] if idx < feat_stack.shape[0] else np.zeros_like(feat_stack[0]))
                    else:
                        ordered_list.append(np.zeros_like(feat_stack[0]))
                ordered = np.stack(ordered_list, axis=0)
            mat = ordered.reshape(ordered.shape[0], -1).T
            return pad_align_matrix(mat)

        def extend(cube):
            if cube is None: return None
            return np.concatenate([cube, cube[-1][None]], axis=0)

        frames=[]; day_summaries=[]; cumulative_pred_mask = np.zeros((Hh,Ww), dtype=bool)

        if SAVE_DAILY_GEOTIFFS and have_rasterio:
            profile = None
            try:
                import rasterio
                if 'stack_files' in globals() and stack_files:
                    with rasterio.open(stack_files[0]) as src:
                        profile = src.profile
                if profile:
                    profile.update(count=1, dtype='float32', compress='deflate')
            except Exception:
                profile = None
        else:
            profile = None

        for step in range(1, FORECAST_DAYS_TO_SHOW+1):
            past_dyn = np.concatenate([past_dyn, last_dyn[None]], axis=0)
            precip_future = extend(precip_future); rh_future=extend(rh_future); vpd_future=extend(vpd_future)
            dryness_future=extend(dryness_future); dry_day_count_future=extend(dry_day_count_future)
            hot_days_future=extend(hot_days_future); windy_days_future=extend(windy_days_future)

            feat_stack_future = build_canonical_feature_stack(past_dyn, step)
            X_future = to_model_matrix(feat_stack_future)
            if lightgbm_mode:
                probs_future = model.predict(X_future)
            else:
                probs_future = model.predict_proba(X_future)[:,1]
            prob_map = probs_future.reshape(Hh, Ww)
            if DEBUG and step==1:
                print('Day+1 prob stats: min', prob_map.min(), 'mean', prob_map.mean(), 'p95', np.percentile(prob_map,95), 'max', prob_map.max())
            if PROB_POWER and PROB_POWER>1:
                prob_map = np.clip(prob_map,0,1)**PROB_POWER
            prob_map_s = smooth(prob_map, SMOOTH_SIGMA)
            high_mask = prob_map_s >= thr
            adaptive_used = False
            if high_mask.sum()==0 and USE_ADAPTIVE_THRESHOLD:
                flat = prob_map_s.ravel()
                k = max(ADAPTIVE_MIN_PIXELS, int(ADAPTIVE_TOP_FRACTION*flat.size))
                k = min(k, flat.size-1)
                if k>0:
                    kth = np.partition(flat, -k)[-k]
                    high_mask = prob_map_s >= kth
                    adaptive_used = True
                    if DEBUG:
                        print(f'Adaptive threshold used for +{step}: {kth:.4f} (top {k} pixels)')
            if FEEDBACK_BURN and BURN_FEEDBACK_MIN_PROB:
                feedback_mask = prob_map_s >= BURN_FEEDBACK_MIN_PROB
                cumulative_pred_mask |= feedback_mask
                last_dyn[5][feedback_mask] = 1

            actual_future=None
            if 'refined_future' in globals() and isinstance(refined_future, np.ndarray):
                idx_future=step-1
                if idx_future < refined_future.shape[0]:
                    cand = refined_future[idx_future]
                    if isinstance(cand,np.ndarray) and cand.shape==(Hh,Ww):
                        actual_future = cand

            if SAVE_DAILY_GEOTIFFS and profile is not None:
                import rasterio
                out_tif = os.path.join(GEOTIFF_DIR, f'prob_day+{step:02d}.tif')
                with rasterio.open(out_tif, 'w', **profile) as dst:
                    dst.write(prob_map.astype('float32'), 1)

            fig, axes = plt.subplots(1,2, figsize=(9,4), dpi=120)
            ax0, ax1 = axes
            im0 = ax0.imshow(prob_map_s, cmap='inferno', vmin=0, vmax=1)
            ax0.set_title(f'Prob Heatmap +{step}')
            ax0.axis('off')
            fig.colorbar(im0, ax=ax0, fraction=0.035, pad=0.02)

            ax1.imshow(prob_map_s, cmap='inferno', vmin=0, vmax=1)
            ax1.imshow(np.ma.masked_where(~high_mask, high_mask), cmap=ListedColormap([[0,0,0,0],[0,1,0,0.35]]), alpha=0.5)
            if cumulative_pred_mask.any():
                hatch_mask = np.ma.masked_where(~cumulative_pred_mask, cumulative_pred_mask)
                ax1.contourf(hatch_mask, levels=[0.5,1], colors='none', hatches=['////'], alpha=0)
            if actual_future is not None:
                ax1.imshow(np.ma.masked_where(actual_future==0, actual_future), cmap=ListedColormap([[0,0,0,0],[0,0,1,0.55]]))
            title_extra = ' (adaptive)' if adaptive_used else ''
            ax1.set_title(f'High-Risk +{step}{title_extra}')
            ax1.axis('off')

            total_pred = int(high_mask.sum())
            total_cum = int(cumulative_pred_mask.sum())
            txt = f"thr={thr:.2f}{'*' if adaptive_used else ''}\nHigh-risk: {total_pred}\nCum spread: {total_cum}\nmin={prob_map.min():.3f} max={prob_map.max():.3f}"
            if actual_future is not None:
                txt += f"\nActual new: {int(actual_future.sum())}"
            ax1.text(0.02,0.02, txt, color='white', fontsize=7, transform=ax1.transAxes,
                     ha='left', va='bottom', path_effects=[patheffects.withStroke(linewidth=2, foreground='black')])
            fut_date = base_date + timedelta(days=step)
            fig.suptitle(f'Forecast Day +{step}  Date: {fut_date.strftime('%Y-%m-%d')}', fontsize=11)
            plt.tight_layout(rect=[0,0,1,0.94])
            buf = io.BytesIO(); fig.savefig(buf, format='png', dpi=120); plt.close(fig); buf.seek(0)
            if pil_ok:
                frames.append(Image.open(buf).convert('P', palette=Image.ADAPTIVE))
            day_summaries.append((step, total_pred, int(actual_future.sum()) if actual_future is not None else None, total_cum, adaptive_used))

        if pil_ok and frames:
            frames[0].save(OUT_GIF, save_all=True, append_images=frames[1:], duration=int(FRAME_DURATION_SEC*1000), loop=0, optimize=False)
            print('Saved enhanced GIF ->', OUT_GIF)
            if WRITE_MP4 and have_imageio:
                try:
                    imgs = [np.array(f.convert('RGB')) for f in frames]
                    imageio.mimsave(OUT_MP4, imgs, fps=int(1/FRAME_DURATION_SEC) if FRAME_DURATION_SEC>0 else 2)
                    print('Saved MP4 ->', OUT_MP4)
                except Exception as e:
                    print('MP4 write failed:', e)
        else:
            print('No frames generated; GIF skipped.')

        print('Day | HighRisk | ActualNew | CumPredSpread | Adaptive')
        for step, hr, act, cum, ada in day_summaries:
            print(f'+{step:02d} | {hr:8d} | {act if act is not None else "  n/a":>9} | {cum:14d} | {"Y" if ada else "N"}')
        print('Green=high-risk, Blue=actual, hatched=cumulative predicted spread, * adaptive threshold used.')
```

```python
# Moisture & Meteorology Feature Assembly (Refactored, Leakage-Safe)
import numpy as np, math, os, importlib
from datetime import datetime, timedelta
from sklearn.model_selection import train_test_split  # (Reserved for calibration split only)

HISTORY_WINDOW_DAYS = 7   # history length (H)
HORIZON_DAYS = 2          # forecast horizon (future burn window)
NEGATIVE_RATIO = 25       # downsample cap (optional)
RANDOM_STATE = 42
USE_LIGHTGBM = True
FORCE_RERUN_MOISTURE = globals().get('FORCE_RERUN_MOISTURE', False)

if 'MOISTURE_RUN_DONE' in globals() and not FORCE_RERUN_MOISTURE:
    print('Moisture feature pipeline already executed. Set FORCE_RERUN_MOISTURE=True to recompute.')
else:
    if 'dynamic_array' not in globals() or 'constants_array' not in globals() or 'dates' not in globals():
        raise RuntimeError('Prerequisites missing: run earlier data assembly cells (dynamic_array, constants_array, dates).')

    T_full, Bdyn, Hh, Ww = dynamic_array.shape
    if T_full <= HISTORY_WINDOW_DAYS + HORIZON_DAYS:
        raise RuntimeError('Not enough temporal depth for chosen history + horizon.')

    # LightGBM availability
    lgbm_available = importlib.util.find_spec('lightgbm') is not None if USE_LIGHTGBM else False
    if lgbm_available:
        import lightgbm as lgb
    else:
        from sklearn.ensemble import RandomForestClassifier
        print('LightGBM not available; falling back to RandomForest.')

    # Indices (assumed ordering from earlier pipeline)
    assumed_band_order = ['TempC','U10','V10','WindSpeed','NDVI','BurnDate']
    idx_temp = assumed_band_order.index('TempC')
    idx_wind = assumed_band_order.index('WindSpeed')
    idx_ndvi = assumed_band_order.index('NDVI')
    idx_burn = assumed_band_order.index('BurnDate')

    temp_all = dynamic_array[:, idx_temp]
    wind_all = dynamic_array[:, idx_wind]
    ndvi_all = dynamic_array[:, idx_ndvi]
    burn_all = dynamic_array[:, idx_burn]   # burn date or indicator layer

    # (Optional) previously retrieved moisture cubes; if absent create zeros
    shape_2d = (T_full, Hh, Ww)
    precip_cube = precip_cube if 'precip_cube' in globals() and isinstance(precip_cube,np.ndarray) and precip_cube.shape == shape_2d else np.zeros(shape_2d, np.float32)
    rh_cube = rh_cube if 'rh_cube' in globals() and isinstance(rh_cube,np.ndarray) and rh_cube.shape == shape_2d else np.zeros(shape_2d, np.float32)
    vpd_cube = vpd_cube if 'vpd_cube' in globals() and isinstance(vpd_cube,np.ndarray) and vpd_cube.shape == shape_2d else np.zeros(shape_2d, np.float32)

    # Day-of-year array
    try:
        doy_list = [datetime.strptime(d, '%Y-%m-%d').timetuple().tm_yday for d in dates]
    except Exception:
        doy_list = list(range(1, T_full+1))
    doy_arr = np.array(doy_list)

    H = HISTORY_WINDOW_DAYS
    i_hist = np.arange(H, dtype=np.float32)
    S_i = i_hist.sum(); S_i2 = (i_hist*i_hist).sum(); DEN = H*S_i2 - S_i**2 if (H*S_i2 - S_i**2)!=0 else 1.0

    def stats(win_rs):
        # win_rs shape: (H, N)
        last = win_rs[-1]
        mean = win_rs.mean(0)
        std = win_rs.std(0)
        Sy = win_rs.sum(0)
        Siy = (win_rs * i_hist[:,None]).sum(0)
        slope = (H * Siy - S_i * Sy)/DEN
        return last, mean, std, slope

    # Track last burn day without future leakage
    last_burn_day = -1 * np.ones(Hh*Ww, dtype=np.int32)

    feature_blocks = []
    label_blocks = []
    day_index_blocks = []  # to enable temporal splitting

    rng = np.random.default_rng(RANDOM_STATE)

    for t in range(H, T_full - HORIZON_DAYS):
        win = slice(t-H, t)           # history days [t-H, t-1]
        future_win = slice(t+1, t+1+HORIZON_DAYS)  # future label window

        # Flatten history windows
        temp_hist = temp_all[win].reshape(H, -1)
        wind_hist = wind_all[win].reshape(H, -1)
        ndvi_hist = ndvi_all[win].reshape(H, -1)
        precip_hist = precip_cube[win].reshape(H, -1)
        rh_hist = rh_cube[win].reshape(H, -1)
        vpd_hist = vpd_cube[win].reshape(H, -1)

        # Stats
        t_last, t_mean, t_std, t_slope = stats(temp_hist)
        w_last, w_mean, w_std, w_slope = stats(wind_hist)
        n_last, n_mean, n_std, n_slope = stats(ndvi_hist)
        p_last, p_mean, p_std, p_slope = stats(precip_hist)
        rh_last, rh_mean, _, _ = stats(rh_hist)
        vpd_last, vpd_mean, _, _ = stats(vpd_hist)

        # Derived features
        t_anom = t_last - t_mean
        n_anom = n_last - n_mean
        dryness = (1 - (n_last + 1)/2).clip(0,1)
        hot_days = (temp_hist > 30.0).sum(0).astype(np.float32)
        windy_days = (wind_hist > 5.0).sum(0).astype(np.float32)
        precip_sum = precip_hist.sum(0)
        dry_day_count = (precip_hist < 0.5).sum(0).astype(np.float32)

        # Days since burn (history only)
        days_since = np.where(last_burn_day >= 0, (t - last_burn_day).astype(np.float32), 0.0)

        # Label: any burn (or active fire if provided) in (t, t+HORIZON] among pixels not yet burned at day t
        burn_today = burn_all[t].reshape(-1) > 0
        future_burn = (burn_all[future_win].reshape(HORIZON_DAYS, -1) > 0).any(axis=0)
        not_burned_yet = ~burn_today  # exclude pixels already burning today
        label_future = future_burn & not_burned_yet

        # (Optional) integrate active fire cube if present to broaden label
        if 'active_fire_cube' in globals() and isinstance(active_fire_cube, np.ndarray) and active_fire_cube.shape[0] >= t+HORIZON_DAYS+1:
            af_future = (active_fire_cube[future_win].reshape(HORIZON_DAYS, -1) > 0).any(axis=0)
            label_future |= (af_future & not_burned_yet)

        # Select valid pixels (exclude currently burning)
        keep_idx = np.where(not_burned_yet)[0]
        if keep_idx.size == 0:
            # Update last_burn_day first then continue
            last_burn_day[burn_today] = t
            continue

        # Day-of-year cyclical features
        doy_t = doy_arr[t]
        sin_doy = math.sin(2*math.pi*doy_t/365.0)
        cos_doy = math.cos(2*math.pi*doy_t/365.0)
        sin_feat = np.full(keep_idx.size, sin_doy, dtype=np.float32)
        cos_feat = np.full(keep_idx.size, cos_doy, dtype=np.float32)

        const_flat = constants_array.reshape(constants_array.shape[0], -1)[:, keep_idx]

        stacked = np.vstack([
            const_flat,
            t_last[keep_idx], t_mean[keep_idx], t_std[keep_idx], t_slope[keep_idx], t_anom[keep_idx],
            w_last[keep_idx], w_mean[keep_idx], w_std[keep_idx], w_slope[keep_idx],
            n_last[keep_idx], n_mean[keep_idx], n_std[keep_idx], n_slope[keep_idx], n_anom[keep_idx], dryness[keep_idx],
            hot_days[keep_idx], windy_days[keep_idx], days_since[keep_idx],
            p_last[keep_idx], p_mean[keep_idx], p_std[keep_idx], p_slope[keep_idx], precip_sum[keep_idx], dry_day_count[keep_idx],
            rh_last[keep_idx], rh_mean[keep_idx], vpd_last[keep_idx], vpd_mean[keep_idx],
            sin_feat, cos_feat
        ])  # NOTE: removed duplicate dry_day_count

        feature_blocks.append(stacked.T)
        label_blocks.append(label_future[keep_idx].astype(np.uint8))
        day_index_blocks.append(np.full(keep_idx.size, t, dtype=np.int32))

        # Update last_burn_day AFTER extracting features/labels (so today burn not included in features)
        last_burn_day[burn_today] = t

    if not feature_blocks:
        raise RuntimeError('No samples generated.')

    X_all = np.concatenate(feature_blocks, axis=0)
    y_all = np.concatenate(label_blocks, axis=0)
    day_idx_all = np.concatenate(day_index_blocks, axis=0)

    print('Assembled samples:', X_all.shape, '| Positives:', int(y_all.sum()), f'Pos%={100*y_all.mean():.4f}')

    # Feature names (align with stacking order)
    FEATURE_NAMES = (
        ['CONST_LULC','CONST_DEM','CONST_Slope','CONST_Aspect','CONST_Hillshade'] +
        ['Temp_last','Temp_mean','Temp_std','Temp_slope','Temp_anom'] +
        ['Wind_last','Wind_mean','Wind_std','Wind_slope'] +
        ['NDVI_last','NDVI_mean','NDVI_std','NDVI_slope','NDVI_anom','Dryness'] +
        ['HotDayCount','WindyDayCount','DaysSinceBurn'] +
        ['Precip_last','Precip_mean','Precip_std','Precip_slope','Precip_sum','DryDayCount'] +
        ['RH_last','RH_mean','VPD_last','VPD_mean'] +
        ['DOY_sin','DOY_cos']
    )
    if X_all.shape[1] != len(FEATURE_NAMES):
        print('WARNING: Feature count mismatch', X_all.shape[1], 'vs names', len(FEATURE_NAMES))

    # Temporal split (keep final 20% of days as test)
    unique_days = np.unique(day_idx_all)
    split_day = unique_days[int(0.8*len(unique_days))]
    train_mask = day_idx_all < split_day
    test_mask = day_idx_all >= split_day

    if train_mask.sum()==0 or test_mask.sum()==0:
        raise RuntimeError('Temporal split failed (empty train or test). Adjust history/horizon or data length.')

    # Downsample negatives in training only (optional)
    POS = np.where(y_all[train_mask]==1)[0]
    NEG = np.where(y_all[train_mask]==0)[0]
    if POS.size and NEGATIVE_RATIO and NEG.size > NEGATIVE_RATIO * POS.size:
        rng = np.random.default_rng(RANDOM_STATE)
        sel_neg = rng.choice(NEG, NEGATIVE_RATIO * POS.size, replace=False)
        sel_idx = np.concatenate([POS, sel_neg])
        rng.shuffle(sel_idx)
        X_train = X_all[train_mask][sel_idx]
        y_train = y_all[train_mask][sel_idx]
        day_train = day_idx_all[train_mask][sel_idx]
        print(f'Downsampled negatives: {NEG.size} -> {sel_neg.size}')
    else:
        X_train = X_all[train_mask]; y_train = y_all[train_mask]; day_train = day_idx_all[train_mask]

    X_test = X_all[test_mask]; y_test = y_all[test_mask]; day_test = day_idx_all[test_mask]

    print('Temporal split: train samples', X_train.shape[0], 'test samples', X_test.shape[0])
    print('Train pos:', int(y_train.sum()), 'Test pos:', int(y_test.sum()))

    # Train model (no leakage)
    if lgbm_available:
        scale_pos = (len(y_train)-y_train.sum())/max(1,y_train.sum()) if y_train.sum()>0 else 1.0
        params = {
            'objective':'binary','learning_rate':0.05,'num_leaves':48,'feature_fraction':0.85,
            'bagging_fraction':0.8,'bagging_freq':1,'min_data_in_leaf':25,'metric':['auc'],
            'verbose':-1,'seed':RANDOM_STATE,'scale_pos_weight':scale_pos
        }
        lgb_train = lgb.Dataset(X_train, label=y_train)
        clf_m = lgb.train(params, lgb_train, num_boost_round=400)
        y_prob = clf_m.predict(X_test)
    else:
        clf_m = RandomForestClassifier(n_estimators=600, max_depth=26, min_samples_leaf=2, n_jobs=-1, random_state=RANDOM_STATE, class_weight='balanced_subsample')
        clf_m.fit(X_train, y_train)
        y_prob = clf_m.predict_proba(X_test)[:,1]

    # Evaluate
    from sklearn.metrics import precision_recall_curve, precision_recall_fscore_support, roc_auc_score, average_precision_score
    if y_test.sum()>0:
        precisions, recalls, thrs = precision_recall_curve(y_test, y_prob)
        f1s = 2*precisions[:-1]*recalls[:-1]/(precisions[:-1]+recalls[:-1]+1e-9)
        best_idx = np.argmax(f1s); BEST_F1_THRESHOLD = thrs[best_idx]
        roc_auc = roc_auc_score(y_test, y_prob) if len(np.unique(y_test))>1 else float('nan')
        pr_auc = average_precision_score(y_test, y_prob)
        print(f'ROC AUC: {roc_auc:.4f} | PR AUC: {pr_auc:.4f}')
        print(f'Best F1 threshold: {BEST_F1_THRESHOLD:.4f} (Prec={precisions[best_idx]:.4f} Rec={recalls[best_idx]:.4f} F1={f1s[best_idx]:.4f})')
    else:
        print('No positives in test set; evaluation skipped.')

    # Expose artifacts
    TEMPORAL_MODEL = clf_m
    TEMP_SPLIT_DAY_INDEX = split_day
    TEMPORAL_FEATURES = FEATURE_NAMES
    labels = y_all  # legacy compatibility
    feature_stacks = None  # not reconstructed here; future cells relying on feature_stacks should adapt.

    MOISTURE_RUN_DONE = True
    print('Leakage-safe moisture feature pipeline complete.')
```

```python
# Permutation importance (optional) â€“ safe, idempotent
import numpy as np
from sklearn.metrics import average_precision_score

def compute_permutation_importance(model=None, feature_stacks=None, labels=None, split_day=None,
                                    max_features=35, random_state=123, top_k=15):
    """Compute permutation importance on temporal test set.
    Requirements: trained model, feature_stacks (T,F,H,W), labels (T or T-1,H,W), split_day index.
    Skips gracefully if any prerequisite missing or inconsistent.
    """
    missing = []
    if model is None: missing.append('model')
    if feature_stacks is None: missing.append('feature_stacks')
    if labels is None: missing.append('labels')
    if split_day is None: missing.append('split_day')
    if missing:
        print('Skip permutation importance â€“ missing:', ', '.join(missing))
        return None
    if not hasattr(model, 'predict'):
        print('Skip â€“ model has no predict method')
        return None
    try:
        T,F,H,W = feature_stacks.shape
    except Exception:
        print('Skip â€“ feature_stacks has unexpected shape')
        return None
    if labels.shape[0] == T:  # label for day t corresponds to features from day t-1
        y_eff = labels[1:]
        feat_eff = feature_stacks[:-1]
    elif labels.shape[0] == T-1:
        y_eff = labels
        feat_eff = feature_stacks[:-1]
    else:
        print('Skip â€“ labels temporal length mismatch.')
        return None
    Td = feat_eff.shape[0]
    HWT = H*W
    # Flatten per-day
    X_flat = feat_eff.reshape(Td, F, HWT).transpose(0,2,1).reshape(Td*HWT, F)
    y_flat = y_eff.reshape(Td, HWT).ravel()
    # Temporal test mask
    day_indices = np.repeat(np.arange(Td), HWT)
    if split_day >= Td:
        print(f'Skip â€“ split_day {split_day} >= available days {Td}')
        return None
    test_mask = day_indices >= split_day
    if not np.any(test_mask):
        print('Skip â€“ no test samples after split_day')
        return None
    X_test = X_flat[test_mask]
    y_test = y_flat[test_mask]
    if y_test.sum() == 0:
        print('Skip â€“ no positive samples in test set')
        return None
    # Base probabilities
    if hasattr(model, 'predict_proba'):
        base_prob = model.predict_proba(X_test)[:,1]
    else:
        base_prob = model.predict(X_test)
    # Calibrator support
    if 'iso' in globals() and globals()['iso'] is not None:
        try:
            base_prob = globals()['iso'].transform(base_prob)
        except Exception:
            pass
    base_pr = average_precision_score(y_test, base_prob)
    feat_names = globals().get('FULL_FEATURE_ORDER', [f'F{i}' for i in range(F)])
    max_features = min(max_features, len(feat_names), F)
    rng = np.random.default_rng(random_state)
    drops = []
    for fi in range(max_features):
        X_perm = X_test.copy()
        col = X_perm[:, fi].copy()
        rng.shuffle(col)
        X_perm[:, fi] = col
        if hasattr(model, 'predict_proba'):
            p_perm = model.predict_proba(X_perm)[:,1]
        else:
            p_perm = model.predict(X_perm)
        if 'iso' in globals() and globals()['iso'] is not None:
            try:
                p_perm = globals()['iso'].transform(p_perm)
            except Exception:
                pass
        pr_perm = average_precision_score(y_test, p_perm)
        drops.append((feat_names[fi], base_pr - pr_perm))
    drops.sort(key=lambda x: x[1], reverse=True)
    print(f'Permutation importance (test samples={y_test.size}, positives={int(y_test.sum())})')
    print(f'Baseline PR AUC: {base_pr:.4f}')
    for name, delta in drops[:top_k]:
        print(f'{name:20s} Î”PR={delta:.5f}')
    return drops

# Auto-run only if user already trained temporal model
if 'TEMP_SPLIT_DAY_INDEX' in globals() and 'feature_stacks' in globals() and 'labels' in globals():
    model_candidate = globals().get('TEMPORAL_MODEL') or globals().get('clf') or globals().get('clf_m')
    compute_permutation_importance(model_candidate, globals().get('feature_stacks'), globals().get('labels'), globals().get('TEMP_SPLIT_DAY_INDEX'))
else:
    print('Permutation importance idle â€“ run temporal training first.')
```

```python
import numpy as np, os, math, rasterio
from rasterio.transform import from_origin
print('Full per-day 35-band GeoTIFF export + full retrain')
needed = ['constants_array','dynamic_array','precip_cube','rh_cube','vpd_cube','active_fire_cube','DATE_LIST']
missing=[n for n in needed if n not in globals()]
if missing:
    print('Missing inputs:', missing)
else:
    const = constants_array.astype('float32')
    dyn = dynamic_array.astype('float32')
    T_full = dyn.shape[0]; H_, W_ = dyn.shape[2], dyn.shape[3]
    precip = precip_cube.astype('float32')
    rhc = rh_cube.astype('float32')
    vpd = vpd_cube.astype('float32')
    fire = active_fire_cube
    crs_out = 'EPSG:4326'
    if 'lon' in globals() and 'lat' in globals():
        west = float(lon[0,0]); east = float(lon[0,-1])
        north = float(lat[0,0]); south = float(lat[-1,0])
        xres = (east - west)/(W_-1) if W_>1 else 0.0001
        yres = abs((south - north)/(H_-1)) if H_>1 else 0.0001
        transform = from_origin(west, north, xres, yres)
    else:
        transform = from_origin(0,0,1,1)
    window = 7
    def window_slice(t):
        s = max(0, t-window+1)
        return slice(s, t+1)
    def rolling_stats(arr3d, t):
        sl = window_slice(t)
        block = arr3d[sl]
        last = arr3d[t]
        mean = block.mean(axis=0)
        std = block.std(axis=0)
        slope = (block[-1]-block[0])/(block.shape[0]-1) if block.shape[0]>1 else np.zeros_like(last)
        anom = last - arr3d[:t+1].mean(axis=0)
        return last, mean, std, slope, anom
    temp_series = dyn[:,0]
    windspeed_series = dyn[:,3]
    ndvi_series = dyn[:,4]
    burn_date_series = dyn[:,-1]
    last_burn_day = np.full((H_,W_), -999, dtype='int16')
    days_since_burn_all = []
    for t in range(T_full):
        burned_today = (fire[t] > 0)
        last_burn_day[burned_today] = t
        days_since = np.where(last_burn_day>=0, t - last_burn_day, np.nan)
        days_since_burn_all.append(days_since.astype('float32'))
    days_since_burn_all = np.stack(days_since_burn_all, axis=0)
    hot_thr = HOT_TEMP_C if 'HOT_TEMP_C' in globals() else 30.0
    windy_thr = WINDY_MS if 'WINDY_MS' in globals() else 5.0
    precip_dry_thr = 1.0
    import datetime
    date_objs = [datetime.datetime.strptime(d,'%Y-%m-%d') for d in DATE_LIST[:T_full]]
    order = FULL_FEATURE_ORDER if 'FULL_FEATURE_ORDER' in globals() else [
        'CONST_LULC','CONST_DEM','CONST_Slope','CONST_Aspect','CONST_Hillshade',
        'Temp_last','Temp_mean','Temp_std','Temp_slope','Temp_anom',
        'Wind_last','Wind_mean','Wind_std','Wind_slope',
        'NDVI_last','NDVI_mean','NDVI_std','NDVI_slope','NDVI_anom',
        'Dryness','HotDayCount','WindyDayCount','DaysSinceBurn','DOY_sin','DOY_cos',
        'Precip_last','Precip_mean','Precip_std','Precip_slope','Precip_sum',
        'RH_last','RH_mean','VPD_last','VPD_mean','DryDayCount'
    ]
    out_dir = 'daily_feature_stacks'
    os.makedirs(out_dir, exist_ok=True)
    X_list = []
    y_list = []
    dates_used = []
    for t in range(T_full):
        temp_last,temp_mean,temp_std,temp_slope,temp_anom = rolling_stats(temp_series, t)
        wind_last,wind_mean,wind_std,wind_slope,_ = rolling_stats(windspeed_series, t)
        ndvi_last,ndvi_mean,ndvi_std,ndvi_slope,ndvi_anom = rolling_stats(ndvi_series, t)
        p_last,p_mean,p_std,p_slope,_ = rolling_stats(precip, t)
        p_sum = precip[window_slice(t)].sum(axis=0)
        rh_last,rh_mean,_,_,_ = rolling_stats(rhc, t)
        vpd_last,vpd_mean,_,_,_ = rolling_stats(vpd, t)
        dryness = vpd_last - rh_last
        hot_day_count = (temp_series[window_slice(t)] > hot_thr).sum(axis=0).astype('float32')
        windy_day_count = (windspeed_series[window_slice(t)] > windy_thr).sum(axis=0).astype('float32')
        dry_day_count = (precip[window_slice(t)] < precip_dry_thr).sum(axis=0).astype('float32')
        days_since_burn = days_since_burn_all[t]
        doy = date_objs[t].timetuple().tm_yday
        sin_doy = np.sin(2*np.pi*doy/365.0)*np.ones((H_,W_),dtype='float32')
        cos_doy = np.cos(2*np.pi*doy/365.0)*np.ones((H_,W_),dtype='float32')
        feature_planes = {
            'CONST_LULC': const[0], 'CONST_DEM': const[1], 'CONST_Slope': const[2], 'CONST_Aspect': const[3], 'CONST_Hillshade': const[4],
            'Temp_last': temp_last, 'Temp_mean': temp_mean, 'Temp_std': temp_std, 'Temp_slope': temp_slope, 'Temp_anom': temp_anom,
            'Wind_last': wind_last, 'Wind_mean': wind_mean, 'Wind_std': wind_std, 'Wind_slope': wind_slope,
            'NDVI_last': ndvi_last, 'NDVI_mean': ndvi_mean, 'NDVI_std': ndvi_std, 'NDVI_slope': ndvi_slope, 'NDVI_anom': ndvi_anom,
            'Dryness': dryness, 'HotDayCount': hot_day_count, 'WindyDayCount': windy_day_count, 'DaysSinceBurn': days_since_burn,
            'DOY_sin': sin_doy, 'DOY_cos': cos_doy,
            'Precip_last': p_last, 'Precip_mean': p_mean, 'Precip_std': p_std, 'Precip_slope': p_slope, 'Precip_sum': p_sum,
            'RH_last': rh_last, 'RH_mean': rh_mean, 'VPD_last': vpd_last, 'VPD_mean': vpd_mean, 'DryDayCount': dry_day_count
        }
        band_stack = np.stack([feature_planes[k] for k in order], axis=0).astype('float32')
        date_str = DATE_LIST[t].replace('-','')
        out_path = os.path.join(out_dir, f'features_{date_str}.tif')
        if not os.path.exists(out_path):
            profile = {'driver':'GTiff','height':H_,'width':W_,'count':band_stack.shape[0],'dtype':'float32','compress':'lzw','transform':transform,'crs':crs_out,'nodata':np.nan}
            with rasterio.open(out_path,'w',**profile) as dst:
                for b in range(band_stack.shape[0]):
                    dst.write(band_stack[b], b+1)
                    dst.set_band_description(b+1, order[b])
        if t < T_full - 1:
            y_map = (fire[t+1] > 0).astype('uint8')
            X_list.append(band_stack.reshape(len(order), -1).T)
            y_list.append(y_map.flatten())
            dates_used.append(DATE_LIST[t])
        if (t+1) % 10 == 0 or t == T_full-1:
            print(f'Processed day {t+1}/{T_full}')
    X_all_new = np.vstack(X_list)
    y_all_new = np.concatenate(y_list)
    print('Assembled training matrix', X_all_new.shape, 'positives', y_all_new.sum())
    from sklearn.ensemble import RandomForestClassifier
    rf = RandomForestClassifier(n_estimators=300, n_jobs=-1, random_state=42, class_weight='balanced_subsample', min_samples_leaf=2)
    rf.fit(X_all_new, y_all_new)
    from sklearn.metrics import classification_report
    preds = rf.predict(X_all_new)
    print(classification_report(y_all_new, preds, digits=4))
    model = rf
    feat_names = order
    X_all = X_all_new; y_all = y_all_new
    print('Export & retrain complete. Daily GeoTIFFs in', out_dir)
```

```python
# Temporal (leak-safe) train/test split + optional spatial blocking
import numpy as np, math
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, precision_recall_curve, roc_auc_score, average_precision_score

print('--- Temporal Split Training (Leakage Mitigation) ---')

# 1. Source feature cube (T,F,H,W) and labels (T-1,H,W) where label[t] = fire at day t+1
feat_src = None
if 'feature_stacks' in globals() and isinstance(feature_stacks, np.ndarray):
    feat_src = feature_stacks  # expected (T,F,H,W)
elif 'feat_cube' in globals() and isinstance(feat_cube, np.ndarray):
    feat_src = feat_cube

if feat_src is None:
    print('Feature cube not found (feature_stacks / feat_cube). Run earlier feature assembly cells.')
else:
    T,Fc,Hc,Wc = feat_src.shape
    if 'labels' not in globals() or not isinstance(labels, np.ndarray):
        print('labels array missing; cannot continue.')
    else:
        # Align lengths: labels should have T-1 along time (label for next day)
        if labels.shape[0] == T:
            # If same length, assume labels[t] already corresponds to day t (we will shift)
            labels_eff = labels[1:]  # drop first to make next-day style
            feat_eff = feat_src[:-1]
            date_offset = 1
        elif labels.shape[0] == T-1:
            labels_eff = labels.copy()
            feat_eff = feat_src[:-1]
            date_offset = 1
        else:
            print('Unexpected labels time dimension', labels.shape[0], 'vs features T', T)
            labels_eff = None
        if labels_eff is not None:
            Td = feat_eff.shape[0]
            print(f'Using {Td} days of features with shape (T={Td},F={Fc},H={Hc},W={Wc})')

            # 2. Build per-pixel samples grouped by day to avoid pixel-level temporal leakage
            day_indices = np.repeat(np.arange(Td), Hc*Wc)
            X_flat = feat_eff.reshape(Td, Fc, Hc*Wc).transpose(0,2,1).reshape(Td*Hc*Wc, Fc)
            y_flat = labels_eff.reshape(Td, Hc*Wc).ravel()

            # Optional: remove clearly leaky bands (e.g., direct burn indicator if present in feature list)
            drop_names = {'DryDayCount'}  # example; extend if needed
            if 'feat_names' in globals() and len(feat_names)==Fc:
                keep_idx = [i for i,n in enumerate(feat_names) if n not in drop_names]
                if len(keep_idx) < Fc:
                    X_use = X_flat[:, keep_idx]
                    kept_names = [feat_names[i] for i in keep_idx]
                    print('Dropped features:', sorted(set(feat_names)-set(kept_names)))
                else:
                    X_use = X_flat; kept_names = feat_names
            else:
                X_use = X_flat; kept_names = ['F'+str(i) for i in range(X_use.shape[1])]

            # 3. Temporal split (e.g., last 20% days for test)
            test_fraction = 0.2
            split_day = int((1-test_fraction)*Td)
            if split_day >= Td-1: split_day = Td-2
            train_mask_days = day_indices < split_day
            test_mask_days  = day_indices >= split_day

            X_train = X_use[train_mask_days]
            y_train = y_flat[train_mask_days]
            X_test  = X_use[test_mask_days]
            y_test  = y_flat[test_mask_days]

            uniq_train_days = np.unique(day_indices[train_mask_days])
            uniq_test_days  = np.unique(day_indices[test_mask_days])
            print('Train days:', uniq_train_days[[0,-1]] if uniq_train_days.size>1 else uniq_train_days, 'count', uniq_train_days.size)
            print('Test days :', uniq_test_days[[0,-1]] if uniq_test_days.size>1 else uniq_test_days, 'count', uniq_test_days.size)
            overlap = set(uniq_train_days).intersection(set(uniq_test_days))
            if overlap:
                print('WARNING: Overlap in day splits ->', overlap)

            # 4. (Optional) Spatial downsample negatives in training to reduce imbalance
            neg_mask = (y_train==0)
            pos_mask = ~neg_mask
            n_pos = pos_mask.sum(); n_neg = neg_mask.sum()
            target_neg = min(n_neg, n_pos*50)  # cap ratio 50:1
            if target_neg < n_neg:
                neg_indices = np.where(neg_mask)[0]
                keep_neg = np.random.default_rng(42).choice(neg_indices, size=target_neg, replace=False)
                keep_all = np.concatenate([np.where(pos_mask)[0], keep_neg])
                X_train = X_train[keep_all]
                y_train = y_train[keep_all]
                print(f'Downsampled train negatives {n_neg}->{target_neg}')
            print('Train shape', X_train.shape, 'Test shape', X_test.shape, 'Pos train', y_train.sum(), 'Pos test', y_test.sum())

            # 5. Fit model
            clf_temp = RandomForestClassifier(n_estimators=300, max_depth=None, n_jobs=-1, class_weight='balanced_subsample', random_state=42)
            clf_temp.fit(X_train, y_train)

            # 6. Evaluate
            if hasattr(clf_temp,'predict_proba'):
                y_prob = clf_temp.predict_proba(X_test)[:,1]
            else:
                y_prob = clf_temp.predict(X_test)
            y_pred_default = (y_prob>=0.5).astype(int)
            print('\nClassification report @0.50 threshold:')
            print(classification_report(y_test, y_pred_default, digits=4))
            try:
                roc = roc_auc_score(y_test, y_prob); pr = average_precision_score(y_test, y_prob)
                print(f'ROC AUC={roc:.4f}  PR AUC={pr:.4f}')
            except Exception: pass

            precisions, recalls, thrs = precision_recall_curve(y_test, y_prob)
            f1s = 2*precisions[:-1]*recalls[:-1]/(precisions[:-1]+recalls[:-1]+1e-9)
            bi = np.argmax(f1s); best_thr = thrs[bi]
            print(f'Best F1 threshold {best_thr:.4f} (P={precisions[bi]:.3f} R={recalls[bi]:.3f} F1={f1s[bi]:.3f})')
            y_pred_best = (y_prob>=best_thr).astype(int)
            print('Report @best F1:')
            print(classification_report(y_test, y_pred_best, digits=4))

            # 7. Store artifacts
            TEMPORAL_MODEL = clf_temp
            TEMPORAL_FEATURES = kept_names
            TEMP_SPLIT_DAY_INDEX = split_day
            TEMPORAL_BEST_F1_THRESHOLD = float(best_thr)
            print('Stored TEMPORAL_MODEL and thresholds.')
print('--- Done Temporal Split ---')
```

```python
# Recompute DaysSinceBurn (lagged: excludes current day burn)
import numpy as np
print('--- Recompute DaysSinceBurn (lag) ---')
if 'dynamic_array' not in globals():
    print('dynamic_array missing; skip.')
elif 'FULL_FEATURE_ORDER' not in globals():
    print('FULL_FEATURE_ORDER missing; skip.')
else:
    try:
        burn_band_idx = 5  # burn date/active burn band in dynamic_array
        burn_hist = (dynamic_array[:, burn_band_idx] > 0).astype(np.uint8)  # (T,H,W)
        T_,H_,W_ = burn_hist.shape
        dsb = np.zeros((T_,H_,W_), dtype=np.float32)
        last_burn = np.full((H_,W_), -1, dtype=np.int32)
        for t in range(T_):
            # days since last burn BEFORE day t
            if t>0:
                # compute using last_burn indices
                mask = last_burn >= 0
                dsb[t] = 0
                dsb[t][mask] = (t - last_burn[mask]).astype(np.float32)
            # update last_burn after recording dsb
            newly_burned = burn_hist[t] > 0
            last_burn[newly_burned] = t
        # Integrate into feature_stacks / feat_cube if present
        target_idx = None
        if 'FULL_FEATURE_ORDER' in globals():
            try:
                target_idx = FULL_FEATURE_ORDER.index('DaysSinceBurn')
            except ValueError:
                pass
        updated = False
        if target_idx is not None:
            if 'feature_stacks' in globals() and isinstance(feature_stacks, np.ndarray) and feature_stacks.shape[0] == T_:
                feature_stacks[:, target_idx] = dsb
                updated = True
            if 'feat_cube' in globals() and isinstance(feat_cube, np.ndarray) and feat_cube.shape[0] == T_:
                feat_cube[:, target_idx] = dsb
                updated = True
        print('Updated DaysSinceBurn band' if updated else 'Did not update (shape mismatch or not found).')
    except Exception as e:
        print('Failed recompute:', e)
print('--- Done ---')
```

```python
# Leakage assertions: temporal boundaries & feature/label alignment
import numpy as np
print('--- Leakage Assertions ---')
issues = []
if 'feature_stacks' in globals() and 'labels' in globals():
    Tfeat = feature_stacks.shape[0]
    Tlab  = labels.shape[0]
    if Tlab == Tfeat:
        print('labels length equals feature days; ensure you shift labels by +1 for prediction.')
    elif Tlab == Tfeat-1:
        print('labels aligned as next-day target (T-1).')
    else:
        issues.append(f'Label length mismatch: features {Tfeat}, labels {Tlab}')
else:
    issues.append('feature_stacks or labels missing.')
# Check DaysSinceBurn band monotonicity (should not decrease unless new burn occurs)
if 'FULL_FEATURE_ORDER' in globals() and 'feature_stacks' in globals():
    if 'DaysSinceBurn' in FULL_FEATURE_ORDER:
        idx = FULL_FEATURE_ORDER.index('DaysSinceBurn')
        dsb_seq = feature_stacks[:, idx]
        # For a few sample pixels
        for r,c in [(0,0),(min(5,dsb_seq.shape[1]-1),min(5,dsb_seq.shape[2]-1))]:
            series = dsb_seq[:, r, c]
            # Skip all zeros / trivial
            if series.max() == 0: continue
            # Check decreases larger than 1 day not associated with reset to 0
            for t in range(2, len(series)):
                if series[t] + 2 < series[t-1] and not (series[t-1]>0 and series[t]==0):
                    issues.append(f'DaysSinceBurn anomaly at t={t} pixel {(r,c)}: {series[t-1]} -> {series[t]}')
                    break

if issues:
    print('ISSUES FOUND:')
    for s in issues: print(' -', s)
else:
    print('No leakage anomalies detected with current checks.')
print('--- End Assertions ---')
```

```python
# Permutation importance on temporal test
import numpy as np
from sklearn.metrics import average_precision_score
print('--- Permutation Importance (Temporal Test) ---')
if 'TEMPORAL_MODEL' not in globals():
    print('TEMPORAL_MODEL missing. Run temporal training cell.')
elif 'TEMP_SPLIT_DAY_INDEX' not in globals() or 'feature_stacks' not in globals() or 'labels' not in globals():
    print('Prerequisites missing (split index / feature_stacks / labels).')
else:
    model = TEMPORAL_MODEL
    T,F,H,W = feature_stacks.shape
    split_day = TEMP_SPLIT_DAY_INDEX
    # Recreate X_test, y_test consistent with temporal cell logic
    if labels.shape[0] == T:  # shift
        y_eff = labels[1:]
        feat_eff = feature_stacks[:-1]
    else:
        y_eff = labels.copy()
        feat_eff = feature_stacks[:-1]
    Td = feat_eff.shape[0]
    day_indices = np.repeat(np.arange(Td), H*W)
    X_flat = feat_eff.reshape(Td, F, H*W).transpose(0,2,1).reshape(Td*H*W,F)
    y_flat = y_eff.reshape(Td, H*W).ravel()
    test_mask = day_indices >= split_day
    X_test_full = X_flat[test_mask]
    y_test = y_flat[test_mask]
    # Use kept feature names if available
    feat_names_pi = globals().get('TEMPORAL_FEATURES', FULL_FEATURE_ORDER if 'FULL_FEATURE_ORDER' in globals() else [f'F{i}' for i in range(F)])
    base_prob = model.predict_proba(X_test_full)[:,1] if hasattr(model,'predict_proba') else model.predict(X_test_full)
    base_pr = average_precision_score(y_test, base_prob)
    print(f'Baseline PR AUC: {base_pr:.4f} (n_test={y_test.size}, positives={y_test.sum()})')
    rng = np.random.default_rng(123)
    max_features = min(len(feat_names_pi), 35)
    importances = []
    for fi in range(max_features):
        X_perm = X_test_full.copy()
        col = X_perm[:, fi].copy()
        rng.shuffle(col)
        X_perm[:, fi] = col
        prob_perm = model.predict_proba(X_perm)[:,1] if hasattr(model,'predict_proba') else model.predict(X_perm)
        pr_perm = average_precision_score(y_test, prob_perm)
        drop = base_pr - pr_perm
        importances.append((feat_names_pi[fi], drop))
    importances.sort(key=lambda x: x[1], reverse=True)
    print('Top 15 features by PR AUC drop:')
    for name, val in importances[:15]:
        print(f'{name:20s}  Î”PR={val:.5f}')
print('--- Done Permutation Importance ---')
```

```python
# Unified Leakage-Safe Spatiotemporal GIF
import numpy as np, math, os, io
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib import patheffects
try:
    from PIL import Image
    pil_ok = True
except Exception:
    pil_ok = False

REQUIRED = ['dynamic_array','constants_array','dates','TEMPORAL_MODEL','BEST_F1_THRESHOLD']
missing = [r for r in REQUIRED if r not in globals()]
if missing:
    print('Cannot build GIF â€“ missing:', missing)
else:
    model = TEMPORAL_MODEL
    threshold = float(BEST_F1_THRESHOLD)
    Hhist = HISTORY_WINDOW_DAYS if 'HISTORY_WINDOW_DAYS' in globals() else 7
    Tfull, Bd, Hh, Ww = dynamic_array.shape
    GIF_FORECAST_DAYS = 5
    OUTPUT_DIR = globals().get('OUTPUT_DIR','.')
    os.makedirs(OUTPUT_DIR, exist_ok=True)

    # Helper stats
    i_hist = np.arange(Hhist, dtype=np.float32)
    S_i = i_hist.sum(); S_i2 = (i_hist*i_hist).sum(); DEN = Hhist*S_i2 - S_i**2 if (Hhist*S_i2 - S_i**2)!=0 else 1.0
    def stats(win_rs):
        last = win_rs[-1]; mean = win_rs.mean(0); std = win_rs.std(0)
        Sy = win_rs.sum(0); Siy = (win_rs * i_hist[:,None,None]).sum(0)
        slope = (Hhist * Siy - S_i * Sy)/DEN if DEN!=0 else np.zeros_like(mean)
        return last, mean, std, slope

    # Slices indices (assumed order)
    idx_temp, idx_wind, idx_ndvi, idx_burn = 0, 3, 4, 5  # based on earlier dynamic_array layout

    # Build historical frames covering days that have at least one full history window
    start_day = Hhist
    frames = []
    burn_series = dynamic_array[:, idx_burn]
    base_dates = [datetime.strptime(d,'%Y-%m-%d') for d in dates]

    # Rolling last burn day to compute DaysSinceBurn comparable to training
    last_burn_day = -1 * np.ones(Hh*Ww, dtype=np.int32)

    # FEATURE_NAMES used for ordering; fallback if absent
    FEATURE_NAMES = globals().get('TEMPORAL_FEATURES', [])

    def build_feature_vector(window, t_index):
        temp = window[:, idx_temp]
        wind = window[:, idx_wind]
        ndvi = window[:, idx_ndvi]
        burn = window[:, idx_burn]
        t_last,t_mean,t_std,t_slope = stats(temp)
        w_last,w_mean,w_std,w_slope = stats(wind)
        n_last,n_mean,n_std,n_slope = stats(ndvi)
        t_anom = t_last - t_mean
        n_anom = n_last - n_mean
        # Days since burn (history only): any burn occurrence indices in history window
        # last_burn_day is maintained outside
        dsb = np.where(last_burn_day>=0, (t_index - last_burn_day).astype(np.float32), 0.0).reshape(Hh,Ww)
        # Placeholder moisture-related (zeros if not present);
        shape = t_last.shape
        p_last=p_mean=p_std=p_slope=p_sum=np.zeros_like(t_last)
        rh_last=rh_mean=vpd_last=vpd_mean=np.zeros_like(t_last)
        dryness = (1 - (n_last + 1)/2).clip(0,1)
        hot_days = (temp > 30).sum(0).astype(np.float32)
        windy_days = (wind > 5).sum(0).astype(np.float32)
        precip_last, precip_mean, precip_std, precip_slope, precip_sum = p_last, p_mean, p_std, p_slope, p_sum
        dry_day_count = (p_last*0).astype(np.float32)
        doy = base_dates[t_index].timetuple().tm_yday if t_index < len(base_dates) else (t_index%365)+1
        sin_doy = math.sin(2*math.pi*doy/365.0); cos_doy = math.cos(2*math.pi*doy/365.0)
        sin_arr = np.full_like(t_last, sin_doy, dtype=np.float32)
        cos_arr = np.full_like(t_last, cos_doy, dtype=np.float32)
        consts = list(constants_array)
        # Assemble in same order as training FEATURE_NAMES if available
        feat_map = {
            'CONST_LULC': consts[0],'CONST_DEM': consts[1],'CONST_Slope': consts[2],'CONST_Aspect': consts[3],'CONST_Hillshade': consts[4],
            'Temp_last': t_last,'Temp_mean': t_mean,'Temp_std': t_std,'Temp_slope': t_slope,'Temp_anom': t_anom,
            'Wind_last': w_last,'Wind_mean': w_mean,'Wind_std': w_std,'Wind_slope': w_slope,
            'NDVI_last': n_last,'NDVI_mean': n_mean,'NDVI_std': n_std,'NDVI_slope': n_slope,'NDVI_anom': n_anom,'Dryness': dryness,
            'HotDayCount': hot_days,'WindyDayCount': windy_days,'DaysSinceBurn': dsb,
            'Precip_last': precip_last,'Precip_mean': precip_mean,'Precip_std': precip_std,'Precip_slope': precip_slope,'Precip_sum': precip_sum,'DryDayCount': dry_day_count,
            'RH_last': rh_last,'RH_mean': rh_mean,'VPD_last': vpd_last,'VPD_mean': vpd_mean,
            'DOY_sin': sin_arr,'DOY_cos': cos_arr
        }
        if FEATURE_NAMES:
            ordered = [feat_map.get(n, np.zeros_like(t_last)) for n in FEATURE_NAMES]
        else:
            ordered = list(feat_map.values())
        arr = np.stack(ordered, axis=0)
        X = arr.reshape(arr.shape[0], -1).T
        return X, dsb

    # Historical frame loop
    for t in range(start_day, Tfull-1):  # t predicts t+1
        window = dynamic_array[t-Hhist:t]
        X_t, dsb = build_feature_vector(window, t)
        if hasattr(model,'predict_proba'):
            probs = model.predict_proba(X_t)[:,1]
        else:
            probs = model.predict(X_t)
        prob_map = probs.reshape(Hh,Ww)
        high_mask = prob_map >= threshold
        today_burn = burn_series[t] > 0
        # Update last_burn_day AFTER using its value
        newly_burned = (burn_series[t] > 0).reshape(-1)
        last_burn_day[newly_burned] = t

        fig, axes = plt.subplots(1,2, figsize=(8,3.0), dpi=120)
        ax0, ax1 = axes
        ax0.imshow(prob_map, cmap='inferno', vmin=0, vmax=1); ax0.set_title(base_dates[t].strftime('%Y-%m-%d'))
        ax0.axis('off')
        ax1.imshow(np.ma.masked_where(~high_mask, high_mask), cmap=ListedColormap([[0,0,0,0],[1,1,0,0.5]]))
        ax1.imshow(np.ma.masked_where(~today_burn, today_burn), cmap=ListedColormap([[0,0,0,0],[0,0,1,0.5]]))
        ax1.set_title('High Risk vs Burn')
        ax1.axis('off')
        ax1.text(0.02,0.02,f"High={int(high_mask.sum())}\nBurn={int(today_burn.sum())}",color='white',fontsize=7,transform=ax1.transAxes,
                 path_effects=[patheffects.withStroke(linewidth=2, foreground='black')])
        buf=io.BytesIO(); fig.tight_layout(); fig.savefig(buf, format='png'); plt.close(fig); buf.seek(0)
        if pil_ok: frames.append(Image.open(buf).convert('P', palette=Image.ADAPTIVE))

    # Forecast frames (autoregressive extension)
    past_dyn = dynamic_array.copy()
    last_dyn = dynamic_array[-1].copy()
    for step in range(1, GIF_FORECAST_DAYS+1):
        past_dyn = np.concatenate([past_dyn, last_dyn[None]], axis=0)
        window = past_dyn[-Hhist:]
        t_index = Tfull - 1 + step
        X_f, dsb = build_feature_vector(window, min(Tfull-1, Tfull-1))  # use last actual date for DOY cycling
        if hasattr(model,'predict_proba'):
            probs_f = model.predict_proba(X_f)[:,1]
        else:
            probs_f = model.predict(X_f)
        prob_map_f = probs_f.reshape(Hh,Ww)
        high_mask = prob_map_f >= threshold
        fig, ax = plt.subplots(figsize=(4,4), dpi=120)
        ax.imshow(prob_map_f, cmap='inferno', vmin=0, vmax=1); ax.set_title(f'Forecast +{step}')
        ax.axis('off')
        buf=io.BytesIO(); fig.tight_layout(); fig.savefig(buf, format='png'); plt.close(fig); buf.seek(0)
        if pil_ok: frames.append(Image.open(buf).convert('P', palette=Image.ADAPTIVE))

    if frames and pil_ok:
        out_path = os.path.join(OUTPUT_DIR, 'unified_spatiotemporal_fire_spread.gif')
        frames[0].save(out_path, save_all=True, append_images=frames[1:], duration=int(800), loop=0)
        print('Saved GIF ->', out_path)
    else:
        print('No frames generated (check pil availability or data).')
```
